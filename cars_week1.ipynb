```python
import pandas as pd
```

```python
df = pd.read_csv('cars.csv', delimiter=',')
df
```

```python
df.info()
```

### Дубликаты

```python
# поиск дубликатов
df.duplicated().sum()
```

```python
# удаление дубликатов
df = df.drop_duplicates()

# Переназначаем индексы строк, удаляем старные индексы
df = df.reset_index(drop=True)

df
```

```python
df.head(6)
```

```python
df.tail(9)
```

### Индексация

```python
# полный столбец в формате Series
df.loc[:3, 'Transmission']
```

```python
# В формате датафрейма
df.loc[:3, ['Transmission', 'Year']]
```

```python
df.loc[:3, 'Make':'Style']
```

```python
# В виде строки
df.loc[3, 'Make':'Style']
```

```python
# в виде татафрейма
df.loc[3:3, 'Make':'Style']
```

```python
# логическое индексирование
df[df['Transmission']=='Manual'].head(5)
```

### Сортировка

```python
# по умолчанию - по возрастанию
df.sort_values(by = 'Price(euro)').tail(5)
```

```python
# по убыванию
df.sort_values(by = 'Year', ascending=False).head(5)
```

### Сохранение в файл

```python
df.to_csv('cars_no_dup.csv', index=False)
```
